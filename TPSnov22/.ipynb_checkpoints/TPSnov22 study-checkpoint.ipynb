### Calabrated Models

```python
y_tr = train_preds.target
preds_df_calibrated = preds_df.copy()
for c in tqdm(train_preds.columns[:-1]):
    # finds calibrating delta
    X_tr=train_preds[c].values.clip(1e-6, 1-1e-6)
    def fun(x):
        return log_loss(y_tr, scipy.special.expit(scipy.special.logit(X_tr) - x))
    delta = scipy.optimize.minimize(fun, x0=1, method='Nelder-Mead', options={'maxiter': 2000})
    delta = delta.x[0]
    
    # fits calibration to all observations
    X_to_calibrate=preds_df[c].values.clip(1e-6, 1-1e-6)
    y_va_pred = scipy.special.expit(scipy.special.logit(X_to_calibrate) - delta)
    preds_df_calibrated[c] = y_va_pred  
```

```python
def preprocess(df, auc_low, auc_up, log_loss_down, logloss_up, add_stats, only_important, top_split, top_gain, feature:list, ):     
    
    # selects only features passed as important_features
    if only_important:
        important_features = list(set(list(top_split.Feature) + list(top_gain.Feature)))
        important_features = sorted(important_features)
        df = df[important_features]
    
    # selects models with AUC greater than auc_low
    models_good_auc= list(metrics_df[metrics_df['AUC'].ge(auc_low)].index)
    df = df.loc[:, df.columns.isin(models_good_auc)].copy()
    
    # selects models with AUC lower than auc_up
    models_good_auc= list(metrics_df[metrics_df['AUC'].le(auc_up)].index)
    df = df.loc[:, df.columns.isin(models_good_auc)].copy()
    
    # selects only models with log loss higher than log_loss_down
    models_good_logloss= list(metrics_df[metrics_df['LogLoss'].ge(log_loss_down)].index)
    df = df.loc[:, df.columns.isin(models_good_logloss)].copy()
    
    # selects only models with log loss lower than logloss_up
    models_good_logloss= list(metrics_df[metrics_df['LogLoss'].le(logloss_up)].index)
    df = df.loc[:, df.columns.isin(models_good_logloss)].copy()
    
    models_columns = df.columns
    
    # Replace model outputs with their inverse
    if 'inv' in feature:
        for c in tqdm(models_columns):
            df[f'{c}'] = 1-df[c] 
    
    # adds descriptive stats of each row
    if add_stats:
        #adds mean and std 
        df['mean'] = df.agg('mean', axis=1)
        df['std'] = df.agg('std', axis=1)
        df['median'] = df.agg('median', axis=1)
        df['max'] = df.agg('max', axis=1)
        df['min'] = df.agg('min', axis=1)
        df['25%'] = df.agg((lambda x: x.quantile(0.25)), axis=1)
        df['75%'] = df.agg((lambda x: x.quantile(0.75)), axis=1)
    
    # adds differences with top 5 important features
    if len(feature)>0:
        top_5_splits = top_split.nlargest(5, 'Value').Feature.values
        top_5_gains = top_gain.nlargest(5, 'Value').Feature.values
        top_list = list(set(list(top_5_splits)+list(top_5_gains)))
        top_list = sorted(top_list)
        for c in tqdm(models_columns):
            for top_model in top_list:
                if 'diff' in feature:
                    df[f'{c}_diff_{top_model}'] = df[c] - df[top_model]  
                if 'div' in feature:
                    df[f'{c}_div_{top_model}'] = df[c] / df[top_model] 
                if 'diff_c' in feature:
                    df[f'{c}_diff_c_{top_model}'] = df[c] - preds_df_calibrated.loc[df.index,top_model] 
                if 'diff_c_inv' in feature:
                    df[f'{c}_diff_inv_c_{top_model}'] = df[c] - (1-preds_df_calibrated.loc[df.index,top_model])
    
    
    return df
```

# Model

```python
def lgbm_cv(X, y, OOF, boosting_type, colsample_bytree, n_splits, random_state):
    MODELS = []
    oof = np.zeros(X.shape[0])
    n_splits = n_splits

    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=random_state)

    for fold, (trn_idx, val_idx) in enumerate(skf.split(X, y)):
        X_train = X.iloc[trn_idx, :]
        y_train = y.loc[trn_idx]
        X_valid = X.iloc[val_idx, :]
        y_valid = y.loc[val_idx]
        model = LGBMClassifier(
            boosting_type = boosting_type,
            n_estimators=5999,
            colsample_bytree=colsample_bytree,
            learning_rate=0.05,
#             num_leaves=60,
            max_depth=4,
#             reg_lambda=2,
            #     reg_alpha=2,
            random_state=42
        )
        model.fit(
            X_train, 
            y_train,
            eval_set=[(X_valid,y_valid),(X_train,y_train)],
            callbacks=[early_stopping(100), log_evaluation(100)]
        )

        oof[val_idx] = model.predict_proba(X_valid)[:, -1]
        logloss = log_loss(y_valid, oof[val_idx])
        MODELS.append(model)

        print(f"{Fore.GREEN}{Style.BRIGHT}fold {fold} log_loss: {logloss:.6f}{Style.RESET_ALL}")

    OOF.append(oof)
    logloss = log_loss(y, oof)
    auc = roc_auc_score(y, oof)
    print(f"{Fore.RED}{Style.BRIGHT}******** log_loss: {logloss:.6f} *********{Style.RESET_ALL}")
    print(f"{Fore.RED}{Style.BRIGHT}******** ROC AUC: {auc:.6f} *********{Style.RESET_ALL}")
    
    return MODELS, OOF
```

```python
def get_features_importance(model, min_splits, min_gain, X_base):
    feature_imp_split = pd.DataFrame({'Value':model.feature_importances_,'Feature':X_base.columns})
    splits = feature_imp_split.sort_values(by='Value')#.nlargest(n_features, columns='Value')

    feature_imp_gain = pd.DataFrame({'Value':model.booster_.feature_importance(importance_type='gain'),'Feature':X_base.columns})
    gains = feature_imp_gain.sort_values(by='Value')#.nlargest(n_features, columns='Value')
    
    top_split = splits[splits.Value.gt(min_splits)]
    top_gain = gains[gains.Value.gt(min_gain)]

    print(f"""
    Important features based on simple LGBM:
    Number of important (Splits>{min_splits}): {top_split.Feature.unique().__len__()}
    Number of important (Gain>{min_gain}): {top_gain.Feature.unique().__len__()}
    """)
    
    return top_split, top_gain
```

```python
## Get important features from basic LGBM with all 5000 models
X = preprocess(X_general, auc_low=0, auc_up=2, log_loss_down=0, logloss_up=1, add_stats=False, only_important = False, top_split=0, top_gain=0, feature=[])
print(f"Shape of data{X.shape}")

MODELS, OOF = lgbm_cv(X,y,OOF, 'gbdt', 0.7, 5, 42)
```
